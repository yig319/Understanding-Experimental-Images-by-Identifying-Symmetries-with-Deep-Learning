```python
symmetry_classes = ['p1', 'p2', 'pm', 'pg', 
                    'cm', 'pmm', 'pmg', 'pgg', 'cmm', 
                 'p4', 'p4m', 'p4g', 'p3', 'p3m1', 'p31m', 'p6', 'p6m']
symmetry_label_dict = {}
for i, k in enumerate(symmetry_classes):
    symmetry_label_dict[k] = i
```

```python
def write_data(file, create_data, i):
    img = plt.imread(file)[:,:,:3]
    if img.dtype != np.int32 or img.dtype != np.uint8:
        img = img*255
    img = img.astype(np.uint8)
    create_data[i] = img
```

```python
path_root = '/scratch/yichen/preprocess-imagenet-17/*'
for group_dir in glob.glob(path_root):
    group = group_dir.split('/')[-1]
    print(group_dir)
    h5_group = h5_file.create_group(group)

    for ds_name_dir in glob.glob(group_dir+'/*'):
        ds_name = ds_name_dir.split('/')[-1]
        print(ds_name_dir)
        h5_ds_name = h5_group.create_group(ds_name)

        for folder_dir in glob.glob(ds_name_dir+'/*'):
            folder = folder_dir.split('/')[-1]
            print(folder_dir)
            
            file_list = []
            labels = []
            
            # fetch file_list for all images, data_size and labels  
            length = 0
            for sym_class_dir in glob.glob(folder_dir+'/*'):
                sym_class = sym_class_dir.split('/')[-1]
                
                sym_file_list = glob.glob(sym_class_dir+'/*')
                file_list += sym_file_list
                
                labels += [symmetry_label_dict[sym_class]]*len(sym_file_list)
                
                
            img_shape = plt.imread(file_list[0])[:,:,:3].shape
            create_data = h5_ds_name.create_dataset(folder+'_data', shape=(len(file_list), *img_shape),
                                                 dtype=np.uint8)
            create_label = h5_ds_name.create_dataset(folder+'_labels', shape=(len(file_list),),
                                                 dtype=np.uint8)
            for i in range(len(file_list)):                
                write_data(file_list[i], create_data, i)
                create_label[i] = labels[i]
```
